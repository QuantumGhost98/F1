```python
import fastf1.plotting
import f1analytics as f1
```

```python
fastf1.plotting.setup_mpl(color_scheme='fastf1')

session_year = 2025
session_name = "Las vegas Grand Prix"

# Load FP2 session
fp2_session_type = 'FP1'
fp2_session = fastf1.get_session(session_year, session_name, fp2_session_type)
fp2_session.load()

# Load Qualifying session (your current session)
q_session_type = 'FP2'
q_session = fastf1.get_session(session_year, session_name, q_session_type)
q_session.load()
```

```python
# Load your sessions
sessions_data = {
    'fp1': {'session': fp2_session, 'session_name': 'Baku GP', 'year': 2025, 'session_type': 'FP1'},
    'fp2': {'session': q_session, 'session_name': 'Baku GP', 'year': 2025, 'session_type': 'FP2'}
}

# Create the analyzer
cross_session_t = f1.cross_session_telemetry.CrossSessionTelemetry(sessions_data)

# Compare laps with custom channels
cross_session_t.compare_laps([
    ('fp1', 'LEC', 'fastest'),
    ('fp2', 'LEC', 'fastest')
], channels=['Speed', 'Throttle', 'Delta'])
```
